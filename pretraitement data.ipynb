```python
import os
import pandas as pd

extract_path = "content/sanad_dataset"
data = []

# Parcourir tous les fichiers texte dans leurs dossiers
for root, dirs, files in os.walk(extract_path):
    for file in files:
        if file.endswith(".txt"):
            file_path = os.path.join(root, file)
            category = os.path.basename(root)  # La catÃ©gorie est le dossier parent

            try:
                # Essayer d'ouvrir en UTF-8
                with open(file_path, "r", encoding="utf-8") as f:
                    text = f.read().strip()

                # Si UTF-8 Ã©choue, essayer UTF-16
                if not text:
                    with open(file_path, "r", encoding="utf-16") as f:
                        text = f.read().strip()

                if text:
                    data.append({"filename": file, "category": category, "text": text})

            except Exception as e:
                print(f"âŒ Erreur en lisant {file}: {e}")

# Convertir en DataFrame
df = pd.DataFrame(data)

# VÃ©rifier s'il y a des donnÃ©es
if not df.empty:
    print(df.head())  # Afficher un aperÃ§u du dataset
    df.to_csv("content/sanad_dataset.csv", index=False)
    print("âœ”ï¸ Fichier CSV sauvegardÃ© : content/sanad_dataset.csv")
else:
    print("âŒ Toujours aucun fichier valide trouvÃ©. VÃ©rifie si les fichiers sont bien lisibles.")
```

```python
import re

# Fonction de nettoyage du texte
def clean_text(text):
    text = re.sub(r'\s+', ' ', text)  # Supprimer les espaces multiples
    text = text.strip()  # Supprimer les espaces en dÃ©but et fin
    return text

# Appliquer le nettoyage
df['text'] = df['text'].apply(clean_text)

# VÃ©rifier les textes les plus longs et les plus courts
print("ğŸ“ Longueur min/max :", df['text'].str.len().min(), "/", df['text'].str.len().max())

# Sauvegarder le dataset nettoyÃ©
df.to_csv("content/sanad_dataset_clean.csv", index=False)
print("âœ”ï¸ Dataset nettoyÃ© sauvegardÃ© : content/sanad_dataset_clean.csv")
```
